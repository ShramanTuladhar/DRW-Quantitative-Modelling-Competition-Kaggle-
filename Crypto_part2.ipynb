# FEATURE ENGINEERING

```python
#Bid-Ask Imbalance (BAI)

df['bid_ask_imbalance'] = (df['bid_qty'] - df['ask_qty']) / (df['bid_qty'] + df['ask_qty'] + 1e-6)
#Measures pressure between buyers and sellers at the top of the book.
#Positive BAI → more bid depth → potential upward price movement.
#Negative BAI → more ask depth → potential sell-side pressure.

#Net Order Flow Imbalance

df['norm_net_order_flow'] = (df['buy_qty'] - df['sell_qty']) / (df['buy_qty'] + df['sell_qty'] + 1e-6)
#Captures real-time demand vs supply from recent order flow.
#Strong net buying can precede price rises; net selling precedes drops.

# Relative Buy/Sell Intensity

df['buy_sell_ratio'] = df['buy_qty'] / (df['sell_qty'] + 1e-6)
#Measures aggressiveness of buyers vs sellers.
#High ratio = strong buying demand → possible price increase.

#Liquidity Imbalance

df['liquidity_imbalance'] = (df['bid_qty'] + df['buy_qty']) - (df['ask_qty'] + df['sell_qty'])
#Combines passive + aggressive orders to indicate liquidity imbalance.
#Positive → potential bullish signal, negative → potential bearish signal.

#Interaction Terms

df['buy_volume_interaction'] = df['buy_qty'] * df['volume']
df['sell_volume_interaction'] = df['sell_qty'] * df['volume']
#Captures nonlinear impact of aggressive trading + liquidity.
#Might indicate true conviction in market direction when high.
```

```python
df['volume_std_10'] = df['volume'].rolling(10).std()
df['volume_rolling_mean_3'] = df['volume'].rolling(window=3).mean()
df['volume_rolling_std_3'] = df['volume'].rolling(window=3).std()
```

```python
epsilon = 1e-6
df['volume_change'] = df['volume'].diff() / (df['volume'].shift(1) + epsilon)
```

```python
df.fillna(0, inplace=True)
```

```python
df= pd.read_parquet("C:\\Users\\shram\\crypto3_df.parquet", engine='pyarrow')
```

```python
filtered_df = df[df['label'].abs() <= 7]
```

```python
filtered_df
```

```python
train = filtered_df.loc[filtered_df.index < '2023-12-31']
train
```

```python
val_df = filtered_df.loc[filtered_df.index > '2023-12-31']
val_df
```

```python
# Step 2: Split features and targets
X_train = train.drop(columns=['label'])
y_train = train['label']

X_val = val_df.drop(columns=['label'])
y_val = val_df['label']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled=scaler.fit_transform(X_val) 
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

pca = PCA()
X_train_pca = pca.fit_transform(X_train_scaled)

# Plot cumulative explained variance to choose number of components
explained_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(explained_variance)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=150)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
```

```python
X_train_pca
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import pearsonr

# Assume X_train, X_val, y_train, y_val are already defined
model = XGBRegressor(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train_pca, y_train)
y_pred = model.predict(X_val_pca)
```

```python
# Pearson correlation (official metric)
corr, _ = pearsonr(y_val, y_pred)
print("Pearson Correlation:", corr)
```

```python
from xgboost import plot_importance

plot_importance(model, max_num_features=20, importance_type='gain', height=0.5)
plt.title("Top 20 Features (by Gain)")
plt.show()
```

```python
# Get raw importance scores
importance_dict = model.get_booster().get_score(importance_type='gain')

# Convert to DataFrame for sorting
importance_df = pd.DataFrame.from_dict(importance_dict, orient='index', columns=['gain'])
importance_df.index.name = 'feature'
importance_df.reset_index(inplace=True)

# Sort by gain ascending (least important first)
least_important = importance_df.sort_values(by='gain', ascending=True)

# View least important features
print(least_important.head(20))  # or whatever number you want
```

```python
from catboost import CatBoostRegressor

model_cb = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    loss_function='RMSE',
    eval_metric='RMSE',
    verbose=100
)

model_cb.fit(X_train_pca, y_train, eval_set=(X_val_pca, y_val), early_stopping_rounds=50)
```

```python
y_pred = model_cb.predict(X_val_pca)
from scipy.stats import pearsonr
corr, _ = pearsonr(y_val, y_pred)
print("📈 Pearson Correlation:", round(corr, 8))
```

```python
y_pred = model.predict(X_val_pca).flatten()  # or use .ravel()
from scipy.stats import pearsonr
corr, _ = pearsonr(y_val, y_pred)
print("📈 Pearson Correlation:", round(corr, 8))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

model_lr = LinearRegression()
model_lr.fit(X_train_pca, y_train)

y_pred = model_lr.predict(X_val_pca) 
corr, _ = pearsonr(y_val, y_pred)

print("📈 Pearson Correlation (Linear):", round(corr, 6))
```

```python
from sklearn.linear_model import Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

# Models to evaluate
models = {
    'Ridge': Ridge(alpha=1.0),
    'ElasticNet': ElasticNet(alpha=0.1, l1_ratio=0.5),
    'RandomForest': RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1)
}

# Results dictionary
results = {}

# Evaluate each model
for name, model in models.items():
    model.fit(X_train_pca, y_train)
    y_pred = model.predict(X_val_pca)

    # Pearson correlation
    corr, _ = pearsonr(y_val, y_pred)
    rmse = mean_squared_error(y_val, y_pred, squared=False)

    results[name] = {'Pearson Corr': round(corr, 6), 'RMSE': round(rmse, 6)}

# Show results
import pandas as pd
results_df = pd.DataFrame(results).T
print(results_df)
```
